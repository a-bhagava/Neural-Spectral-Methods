```python
# ...existing code...

import numpy as np
import matplotlib.pyplot as plt

def compute_mse_stats(preds, truths, sample_indices=None):
    preds = np.asarray(preds)
    truths = np.asarray(truths)
    # normalize shapes
    if preds.ndim == 3:
        preds = preds[np.newaxis, ...]
    if truths.ndim == 3:
        truths = truths[np.newaxis, ...]
    if preds.ndim == 5 and preds.shape[-1] == 1:
        preds = preds[..., 0]
    if truths.ndim == 5 and truths.shape[-1] == 1:
        truths = truths[..., 0]
    if sample_indices is not None:
        preds = preds[sample_indices]
        truths = truths[sample_indices]
    assert preds.shape == truths.shape, f"{preds.shape} vs {truths.shape}"
    N, T = preds.shape[0], preds.shape[1]
    mse = np.mean((preds - truths) ** 2, axis=(2, 3))  # (N, T)
    mean_mse = mse.mean(axis=0)
    std_mse = mse.std(axis=0, ddof=0)
    stderr = std_mse / np.sqrt(max(1, N))
    ci = 1.96 * stderr
    return {"mean_mse": mean_mse, "ci": ci}

# compute stats for both datasets (these variables exist earlier in the notebook)
# [`data_low_absolute`](notebooks/curledwake_hyperparam_sweeps.ipynb),
# [`data_low_residual`](notebooks/curledwake_hyperparam_sweeps.ipynb),
# [`true_data_easy`](notebooks/curledwake_hyperparam_sweeps.ipynb)
stats_a = compute_mse_stats(data_low_absolute, true_data_easy)
stats_b = compute_mse_stats(data_low_residual, true_data_easy)

xs = np.arange(len(stats_a["mean_mse"]))
fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(xs, stats_a["mean_mse"], label="Low-absolute mean MSE", color="C0")
ax.fill_between(xs,
                stats_a["mean_mse"] - stats_a["ci"],
                stats_a["mean_mse"] + stats_a["ci"],
                color="C0", alpha=0.25, label="Low-absolute 95% CI")

ax.plot(xs, stats_b["mean_mse"], label="Low-residual mean MSE", color="C1")
ax.fill_between(xs,
                stats_b["mean_mse"] - stats_b["ci"],
                stats_b["mean_mse"] + stats_b["ci"],
                color="C1", alpha=0.25, label="Low-residual 95% CI")

ax.set_xlabel("time / snapshot index")
ax.set_ylabel("MSE")
ax.set_title("Comparison: Low-absolute vs Low-residual (mean ± 95% CI)")
ax.grid(alpha=0.3)
ax.legend()
plt.tight_layout()
plt.show()

# ...existing code...
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_mse_summary(preds, truths, sample_indices=None, title="MSE vs time", time=None, ax=None):
    """
    preds: np.array of shape (N, T, Ny, Nz[, C]) or (T, Ny, Nz[, C]) or (T, Ny, Nz)
    truths: same shape as preds (if preds is a subset, truths must match same first-dim ordering)
    sample_indices: optional list/array of trajectory indices to include (first axis)
    time: optional 1D array of length T for x-axis
    returns: dict with mean_mse, std_mse, ci
    """
    preds = np.asarray(preds)
    truths = np.asarray(truths)

    # normalize shapes to (N, T, Ny, Nz)
    if preds.ndim == 3:  # (T, Ny, Nz) -> (1, T, Ny, Nz)
        preds = preds[np.newaxis, ...]
    if truths.ndim == 3:
        truths = truths[np.newaxis, ...]
    # remove trailing channel dim if present
    if preds.ndim == 5 and preds.shape[-1] == 1:
        preds = preds[..., 0]
    if truths.ndim == 5 and truths.shape[-1] == 1:
        truths = truths[..., 0]

    # choose subset of trajectories if requested
    if sample_indices is not None:
        preds = preds[sample_indices]
        truths = truths[sample_indices]

    # sanity shapes
    assert preds.shape == truths.shape, f"preds {preds.shape} vs truths {truths.shape}"

    N, T = preds.shape[0], preds.shape[1]

    # compute MSE per trajectory per time: mean over spatial dims
    mse = np.mean((preds - truths) ** 2, axis=(2, 3))  # shape (N, T)

    mean_mse = mse.mean(axis=0)
    std_mse = mse.std(axis=0, ddof=0)
    stderr = std_mse / np.sqrt(max(1, N))
    ci = 1.96 * stderr  # approximate 95% CI

    if ax is None:
        fig, ax = plt.subplots(figsize=(7, 4))

    xs = time if time is not None else np.arange(T)
    ax.plot(xs, mean_mse, label="mean MSE")
    ax.fill_between(xs, mean_mse - ci, mean_mse + ci, alpha=0.25, label="95% CI")
    ax.set_xlabel("time / snapshot index")
    ax.set_ylabel("MSE")
    ax.set_title(title)
    ax.grid(alpha=0.3)
    ax.legend()

    return {"mse_per_traj": mse, "mean_mse": mean_mse, "std_mse": std_mse, "ci": ci, "N": N}


res = plot_mse_summary(data_low_absolute, true_data_hard, title="Low-absolute dataset MSE (mean ± 95% CI)")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Load wake data
true_data_easy = np.load('/mnt/home/abhagava/Neural-Spectral-Methods/src/pde/curledwake/_u_full_hard.npy')

# pick one trajectory
param_data = true_data_easy[5]   # shape (51, 129,129,1)

# indices to visualize
sample_indices = np.array([0, 15, 25, 48])

# band definitions (frequency radii)
low_cut  = 4      # keep only very low spatial frequencies
high_cut = 15     # keep only high frequencies
# (You can tune these.)

def bandpass_2d(field, low=None, high=None):
    """
    field: 2D array (Nx, Ny)
    low, high: radial frequency bounds (in FFT index units)
    """
    Nx, Ny = field.shape
    # 2D FFT
    F = np.fft.fftshift(np.fft.fft2(field))
    
    # frequency grids
    kx = np.fft.fftshift(np.fft.fftfreq(Nx)) * Nx
    ky = np.fft.fftshift(np.fft.fftfreq(Ny)) * Ny
    KX, KY = np.meshgrid(kx, ky, indexing="ij")
    KR = np.sqrt(KX**2 + KY**2)

    # build mask
    mask = np.ones_like(KR, dtype=bool)
    if low is not None:
        mask &= (KR >= low)
    if high is not None:
        mask &= (KR <= high)

    F_filtered = F * mask
    filtered_field = np.real(np.fft.ifft2(np.fft.ifftshift(F_filtered)))
    return filtered_field

# ------------------------------------------------------------
# Plot original + low-pass + bandpass + high-pass for samples
# ------------------------------------------------------------
fig, axes = plt.subplots(len(sample_indices), 4, figsize=(16, 3*len(sample_indices)))

for row, idx in enumerate(sample_indices):
    field = param_data[idx, :, :, 0]

    lowpass  = bandpass_2d(field, low=None, high=low_cut)
    band     = bandpass_2d(field, low=low_cut, high=high_cut)
    highpass = bandpass_2d(field, low=high_cut, high=None)

    for col, (img, title) in enumerate([
        (field,     "Original"),
        (lowpass,   f"Low-pass (≤{low_cut})"),
        (band,      f"Band {low_cut}–{high_cut}"),
        (highpass,  f"High-pass (≥{high_cut})"),
    ]):
        ax = axes[row, col]
        im = ax.imshow(img, cmap="viridis", origin="lower", extent=[0,1,0,1])
        ax.set_title(f"{title} @ t={idx}")
        plt.colorbar(im, ax=ax)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ----------------------------------------------------
# Load wake dataset
# ----------------------------------------------------
true_data_easy = np.load('/mnt/home/abhagava/Neural-Spectral-Methods/src/pde/curledwake/_u_full_hard.npy')
param_data = true_data_easy[5]   # (51, 129,129,1)
T, Nx, Ny, _ = param_data.shape
M = Nx * Ny

# ----------------------------------------------------
# Build snapshot matrix: U (space_dim × T)
# ----------------------------------------------------
U = param_data[..., 0].reshape(T, M).T   # (M, T)
print("Snapshot matrix U shape:", U.shape)

# Subtract temporal mean (standard POD convention)
U_mean = U.mean(axis=1, keepdims=True)
Uc = U - U_mean

# ----------------------------------------------------
# Compute POD (via SVD)
# ----------------------------------------------------
# Instead of SVD on Uc directly (which is 16641×51), compute eigen-decomposition of C = Ucᵀ Uc (51×51)
C = Uc.T @ Uc
eigvals, eigvecs = np.linalg.eigh(C)        # ascending order
idx = np.argsort(eigvals)[::-1]             # descending
eigvals = eigvals[idx]
eigvecs = eigvecs[:, idx]

# POD spatial modes
Phi = Uc @ (eigvecs / np.sqrt(eigvals)[None, :])   # shape: (M, T)
E = eigvals                                         # modal energies

print("POD modes shape:", Phi.shape)
print("Energy spectrum:", E[:5])

# ----------------------------------------------------
# Plot POD energy spectrum
# ----------------------------------------------------
plt.figure(figsize=(6,4))
plt.semilogy(E / np.sum(E), 'o-')
plt.xlabel("Mode index")
plt.ylabel("Normalized energy")
plt.title("POD Modal Energy Spectrum")
plt.grid(True)
plt.tight_layout()
plt.show()

# ----------------------------------------------------
# Plot first few POD spatial modes
# ----------------------------------------------------
num_modes_to_plot = 4
fig, axes = plt.subplots(1, num_modes_to_plot, figsize=(16,4))

for k in range(num_modes_to_plot):
    mode = Phi[:, k].reshape(Nx, Ny)
    im = axes[k].imshow(mode, cmap='seismic', origin='lower', extent=[0,1,0,1])
    axes[k].set_title(f"POD Mode {k+1}")
    plt.colorbar(im, ax=axes[k])

plt.tight_layout()
plt.show()
```
